```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
df = pd.read_csv('fills.csv')

@interact_manual
def show_articles_more_than(column=['price', 'size'], 
                            x=(10, 20000, 10)):
    return df.loc[df[column] > x]
```
